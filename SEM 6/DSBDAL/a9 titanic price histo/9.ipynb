**Heat Maps**

```python
dataset = sns.load_dataset("titanic")
dataset.head()
```

```python
# corr = dataset.corr()
# sns.heatmap(corr)

# Select numeric columns only
numeric_columns = dataset.select_dtypes(include=[np.number])

# Convert categorical columns to one-hot encoded representation
categorical_columns = dataset.select_dtypes(include=['category', 'object'])
encoded_columns = pd.get_dummies(categorical_columns)

# Concatenate the numeric and encoded categorical columns
processed_dataset = pd.concat([numeric_columns, encoded_columns], axis=1)

# Calculate the correlation matrix
corr = processed_dataset.corr()

# Plot the correlation heatmap
sns.heatmap(corr, annot=True, cmap='coolwarm')
```

```python
import pandas as pd
import seaborn as sns

# Load the Titanic dataset
dataset = sns.load_dataset('titanic')

# Plot a histogram of ticket prices
sns.histplot(dataset['fare'], kde=False)

# Set the plot title and labels
plt.title('Ticket Price Distribution')
plt.xlabel('Ticket Price')
plt.ylabel('Frequency')

# Display the histogram
plt.show()
```
